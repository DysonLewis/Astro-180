```python
import numpy as np
```

### a) Calculate the Fried parameter r_0 at 500 nm given seeing of 0.5 arcseconds
### The relationship between seeing (FWHM) and Fried parameter is:
### seeing (arcsec) ≈ 0.98 * lambda / r_0

```python
lam = 500e-9              # wavelength 500 nm
FWHM_arcsec = 0.5
FWHM_rad = np.deg2rad(FWHM_arcsec/3600.0)  # convert arcsec to radians
r0 = 0.98 * lam / FWHM_rad
print("Fried parameter r0 at 500 nm: {:.3f} m ({:.1f} cm)".format(r0, r0*100))
```

### b) Calculate r_0 at 1.65 microns using Kolmogorov turbulence scaling
### For Kolmogorov turbulence: r_0 ∝ lambda^(6/5)
### Therefore: r_0(lambda_2) = r_0(lambda_1) * (lambda_2 / lambda_1)^(6/5)

```python
lam_ir = 1.65e-6  # wavelength 1.65 microns
r0_ir = r0 * (lam_ir/lam)**(6/5)
print("Fried parameter r0 at 1.65 μm: {:.3f} m ({:.1f} cm)".format(r0_ir, r0_ir*100))
```

### c) Calculate number of sub-apertures for a 10 m telescope
### Number of sub-apertures ≈ (D / r_0)^2
### where D is the telescope diameter

```python
D = 10.0  # telescope diameter in meters
subapertures = (D / r0_ir)**2
print("Number of sub-apertures (approx): {:.0f}".format(subapertures))
```
